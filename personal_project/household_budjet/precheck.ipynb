```python
import requests

url = 'https://open.er-api.com/v6/latest/USD'

response = requests.get(url)
print(f"{response.status_code}")
print(response.json()) # method 타입으로 값이 들어오기 때문에 ()필요 => dict 타입으로 옴

data = response.json()
print(data['rates'])
# print( f"USD -> KRW : {data['rates']['KRW']}" )
# print( f"USD -> JPY : {data['rates']['JPY']}" )
# print( f"USD -> EUR : {data['rates']['EUR']}" )

currencies = ['KRW', 'JPY', 'EUR']
for c in currencies:
    print( f"USD -> {c} : {data['rates'][c]}" )
```

```python
import mysql.connector
from mysql.connector import Error
from dotenv import load_dotenv
import os

load_dotenv()

DB_CONFIG = {
    'host' : os.getenv('HOST', 'localhost'),
    'user' : os.getenv('USER', 'root'),
    'password' : os.getenv('PASSWORD'),
    'port' : os.getenv('PORT', 3306),
    'database' : 'household_budget'
}

class DBmanager:
    def __init__(self, host, user, password, port, database):
        self.host = host
        self.user = user
        self.password = password
        self.port = port
        self.database = database
        self.cursor = None
        self.conn = None

    def DBconnect(self):
        try:
            self.conn = mysql.connector.connect(
                host = self.host,
                user = self.user,
                password = self.password,
                port = self.port,
                database = self.database
            )
            self.cursor = self.conn.cursor()
            print("연결 성공")
        except Exception as e:
            print(f"연결 실패 : {e}")

    def DBclose(self):
        if self.cursor:
            self.cursor.close()
        if self.conn and self.conn.is_connected():
            self.conn.close()
            print(" 연결 종료")
```

```python
db = DBmanager(**DB_CONFIG)

db.DBconnect()
db.DBclose()
```
