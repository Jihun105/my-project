```python
# csv 파일 DB에 넣기

import os
import mysql.connector
from mysql.connector import Error
from dotenv import load_dotenv
from db_manager import DBManager
import pandas as pd
import glob # 특정 패턴의 파일 경로를 한 번에 찾아주는 라이브러리

# raw data DB에 일단 넣는 함수(전처리 나중에)
#===================================================================
def create_table(df, table_name):
    data_mapping = {
        'int64' : 'INT',
        'float64' : 'FLOAT',
        'object' : "VARCHAR(20)"
    }

    columns = []
    for col, dtype in df.dtypes.items():
        sql_type = data_mapping.get(str(dtype), 'VARCHAR(30)')
        columns.append(f" `{col}` {sql_type}")
    columns_str = ',\n'.join(columns)
    
    return f'CREATE TABLE IF NOT EXISTS {table_name} (\n{columns_str}\n);'
#===================================================================
def insert_data(df, table_name):

    columns = list(df.columns)
    columns_str = ', '.join([f'`{col}`' for col in columns])
    
    values = ['%s' for i in range(len(columns))]
    values_str = ', '.join(['%s'] * len(columns))

    return f"INSERT INTO {table_name} ({columns_str}) \nVALUES ({values_str})"


# files = glob.glob('data/data_csv/*.csv')

# for file in files:
#     df = pd.read_csv(file) # csv 파일을 DataFrame으로 불러오기
#     print(f"\n===={file}=====")
#     print(df.dtypes) # 각 컬럼명과 타입 출력
df_team = pd.read_csv('../data/data_csv/team.csv', encoding='utf-8')
df_batter_offence = pd.read_csv('../data/data_csv/batter_offence.csv', encoding='utf-8')
df_batter_defence = pd.read_csv('../data/data_csv/batter_defence.csv',encoding='utf-8')
df_pitcher = pd.read_csv('../data/data_csv/pitcher.csv', encoding='utf-8')


load_dotenv()

DB_CONFIG = {
    'host' : os.getenv('HOST', 'localhost'),
    'user' : os.getenv('USER', 'root'),
    'password' : os.getenv('PASSWORD'),
    'port' : os.getenv('PORT', 3306),
    'database' : os.getenv('DATABASE')
}

db = DBManager(**DB_CONFIG)
db.DBconnect()

db.DBexecute(f"""{create_table(df_team, 'team')}""")
db.DBexecute(f"""{create_table(df_batter_offence, 'batter_offence')}""")
db.DBexecute(f"""{create_table(df_batter_defence, 'batter_defence')}""")
db.DBexecute(f"""{create_table(df_pitcher, 'pitcher')}""")

query = insert_data(df_team, 'team')
for col, row in df_team.iterrows():
    db.DBexecute(query, tuple(row))
query = insert_data(df_batter_offence, 'batter_offence')
for col, row in df_batter_offence.iterrows():
    db.DBexecute(query, tuple(row))
query = insert_data(df_batter_defence, 'batter_defence')
for col, row in df_batter_defence.iterrows():
    db.DBexecute(query, tuple(row))
query = insert_data(df_pitcher, 'pitcher')
for col, row in df_pitcher.iterrows():
    db.DBexecute(query, tuple(row))


db.DBclose()
```
